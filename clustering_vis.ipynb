```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

for i in range(5):
    file_path = f'storage/embeddings/{i}_feats.csv'  # Replace with your file path
    data = pd.read_csv(file_path)
    data_features = data.drop(data.columns[0], axis=1)
    data_features
    weight_data = pd.read_csv(f'storage/embeddings/{i}_sample_weight.csv')
    weights = weight_data.drop(weight_data.columns[0], axis=1)
    weights
    cluster_label_data = pd.read_csv(f'storage/embeddings/{i}_kmeans_labels.csv')
    cluster_labels = cluster_label_data.drop(cluster_label_data.columns[0], axis=1)
    cluster_labels

    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(data_features)

    closest_data = pd.read_csv(f'storage/embeddings/{i}_chosen_idx.csv')
    closest = closest_data.drop(closest_data.columns[0], axis=1).values
    closest_mask = np.zeros(len(data_features), dtype=bool)
    closest_mask[closest] = True
    closest_mask

    pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
    pca_df['Cluster'] = cluster_labels
    pca_df['weight'] = weights
    weight_closest = weights
    weight_closest[closest_mask] = 30
    pca_df['weight+closest'] = weight_closest
    # markers = ['X' if closest_mask[i] else 'o' for i in range(len(data_features))]
    pca_df['anchor'] = ['new anchors' if closest_mask[i] else 'others' for i in range(len(data_features))]

    plt.figure(figsize=(10, 8))
    # create a subplot with 1 row and 2 columns
    # first plot
    fig, ax = plt.subplots(1, 2, figsize=(20, 8))
    sns.scatterplot(x="PC1", y="PC2", hue="Cluster", size="weight", data=pca_df, palette=sns.color_palette("hsv", cluster_labels.nunique().values[0]), ax=ax[0])#
    sns.scatterplot(x="PC1", y="PC2", hue="Cluster", size="weight+closest", data=pca_df, palette=sns.color_palette("hsv", cluster_labels.nunique().values[0]), style='anchor',
    markers={'new anchors': 'X', 'others': 'o'}, ax=ax[1])#
    # c_plot = sns.scatterplot(x="PC1", y="PC2", hue="Cluster", data=pca_df[closest_mask], palette=sns.color_palette("hsv", cluster_labels.nunique().values[0]), marker='X', s=weights.max() * 3)
    # Remove the legend for the second plot
    # c_plot.legend_.remove()
    ax[0].title.set_text('Clusters visualization')
    ax[1].title.set_text('Clusters visualization with new selected anchor')
    plt.show()
```
